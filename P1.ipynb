```python
plt.figure(figsize=(20, 15))
sns.countplot(data=df[:500], x='category')

plt.xlabel('Nominee Category')
plt.ylabel('Count')
plt.title('Number of Nominees by Category')

plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(20, 15))
sns.kdeplot(data=df, x='year_film')

plt.xlabel('Year of film')
plt.ylabel('Count')
plt.title('Estimation of Nominations by Year')

plt.show()
```
